```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Get the raw URL for the CSV file
raw_csv_url = 'https://raw.githubusercontent.com/abdelDebug/Breast-Cancer-Dataset/main/breast_cancer_dataset.csv'

# Read the CSV file using the raw URL
df = pd.read_csv(raw_csv_url)
```

```python
# Drop non-numeric columns for visualizations
df['diagnosis_bin'] = df['diagnosis'].map({'M': 1, 'B': 0})
df = df.drop(columns=['diagnosis'])
numeric_df = df.select_dtypes(include=[np.number])

# Histograms
numeric_df.hist(bins=20, figsize=(16, 12), color='skyblue')
plt.suptitle("Histograms of Numeric Features")
plt.tight_layout()
plt.show()

```

```python
# Target column: 'diagnosis' (we already mapped it to binary as 'diagnosis_bin')
target = 'diagnosis_bin'
X = df.drop(columns=[target])
y = df[target]
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform features
X_scaled = scaler.fit_transform(X)

# Optional: convert back to DataFrame for readability
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Save processed data to CSV
processed_df = pd.concat([X_scaled, y.reset_index(drop=True)], axis=1)
processed_df.to_csv('processed_breast_cancer_data.csv', index=False)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.model_selection import train_test_split

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Training samples: {X_train.shape[0]}, Test samples: {X_test.shape[0]}")
```

```python
# get a list of models to evaluate
def get_models():
	models = dict()
	models['lr'] = LogisticRegression(max_iter=1000, random_state=42)
	models['knn'] = KNeighborsClassifier()
	models['cart'] = DecisionTreeClassifier(random_state=42)
	models['svm'] = SVC(probability=True, random_state=42)
	models['RDF'] = RandomForestClassifier(random_state=42)
	return models
```

```python
# Train each model and store it
trained_models = {}
models = get_models()
for name, model in models.items():
    model.fit(X_train, y_train)
    trained_models[name] = model
    print(f"{name} trained successfully.")
```

```python
for name, model in trained_models.items():
    print(f"\n{name} Hyperparameters:")
    print(model.get_params())
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, roc_curve, RocCurveDisplay
```

```python

# Function to evaluate model performance
def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None

    return {
        'Model': name,
        'Accuracy': accuracy_score(y_test, y_pred),
        'Precision': precision_score(y_test, y_pred),
        'Recall': recall_score(y_test, y_pred),
        'F1 Score': f1_score(y_test, y_pred),
        'AUC': roc_auc_score(y_test, y_proba) if y_proba is not None else "N/A"
    }

# Evaluate all models
results = [evaluate_model(name, model, X_test, y_test) for name, model in trained_models.items()]

# Create summary table
results_df = pd.DataFrame(results)
print(results_df)
```

```python
# Plot confusion matrices
fig, axes = plt.subplots(2,3, figsize = (8,4))
axes = axes.flatten() # Flatten the 2x2 grid of axes into a 1D array for easy iteration
for i, (name, model) in enumerate(trained_models.items()):
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    # Create the ConfusionMatrixDisplay object and plot it on the corresponding axis
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1]) # Add display labels
    disp.plot(ax=axes[i]) # Pass the current axis from the flattened array
    axes[i].set_title(f'{name} - Confusion Matrix') # Set the title on the current axis
plt.tight_layout()
plt.show()

# Plot ROC curves
fig, axes = plt.subplots(2,3, figsize = (8,4))
axes = axes.flatten()
for i, (name, model) in enumerate(trained_models.items()):
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)[:, 1]
        fpr, tpr, _ = roc_curve(y_test, y_proba)
        RocCurveDisplay(fpr=fpr, tpr=tpr).plot(ax=axes[i])
        axes[i].set_title(f'{name} - ROC Curve')

plt.tight_layout()
plt.show()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid_dt = {
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 3, 4],
    'min_samples_leaf': [1, 2],
    'criterion': ['gini', 'entropy'],
    'class_weight': [None, 'balanced']  # New: Helps improve recall
}
grid_search_dt = GridSearchCV(
    estimator=DecisionTreeClassifier(random_state=42),
    param_grid=param_grid_dt,
    scoring='f1',
    cv=5,
    n_jobs=-1
)
```

```python
# Fit model
grid_search_dt.fit(X_train, y_train)

# Best tuned model
dt_tuned = grid_search_dt.best_estimator_
print("Best Parameters:", grid_search_dt.best_params_)
```

```python
dt_tuned = evaluate_model("Decision Tree (Tuned)", dt_tuned, X_test, y_test)
dt_post = evaluate_model("Decision Tree (Default)", trained_models['cart'], X_test, y_test)
```

```python
comparison_df_dt = pd.DataFrame([dt_post,dt_tuned])
print(comparison_df_dt)
```

```python
# Bar plot for visual comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']
comparison_df_dt.set_index('Model')[metrics].plot(kind='bar', figsize=(10, 6))
plt.title("Decision Tree: Pre- vs Post-Tuning Performance")
plt.ylabel("Score")
plt.ylim(0.85, 1.01)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python
#Define parameter grid
param_grid = {
   'C': [0.1, 1, 10, 100],
    'gamma': [1, 0.1, 0.01, 0.001],
    'kernel': ['rbf'],
    'class_weight': [None, 'balanced']
}

# Set up GridSearchCV
svm = SVC(probability=True, random_state=42)

grid_search_svm = GridSearchCV(
    estimator=svm,
    param_grid=param_grid,
    scoring='f1',
    cv=5,
    n_jobs=-1
)
```

```python

# Fit on training data
grid_search_svm.fit(X_train, y_train)

# Get best model
svm_tuned = grid_search_svm.best_estimator_
print("Best Parameters:", grid_search_svm.best_params_)
```

```python
# Evaluate the tuned model
svm_post = evaluate_model("SVM (Tuned)", svm_tuned, X_test, y_test)

svm_default = trained_models['svm']
svm_pre = evaluate_model("SVM (Default)", svm_default, X_test, y_test)
```

```python
comparison_df_dt = pd.DataFrame([svm_pre,svm_post])
print(comparison_df_dt)
```

```python
# Bar plot for visual comparison
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score', 'AUC']
comparison_df_dt.set_index('Model')[metrics].plot(kind='bar', figsize=(10, 6))
plt.title("SVM: Pre- vs Post-Tuning Performance")
plt.ylabel("Score")
plt.ylim(0.85, 1.01)
plt.xticks(rotation=0)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
